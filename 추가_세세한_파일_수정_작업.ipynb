```python
def remove_tags(tags, tags_to_remove):
    # 태그들을 쉼표(,)로 분리하여 리스트로 변환합니다.
    tag_list = tags.split(",")
    # 삭제할 태그들을 리스트로 변환합니다.
    tags_to_remove_list = tags_to_remove.split(",")
    # 삭제할 태그들을 제외한 나머지 태그들을 유지합니다.
    updated_tag_list = [tag for tag in tag_list if tag not in tags_to_remove_list]
    # 남은 태그들을 다시 쉼표로 연결하여 하나의 문자열로 반환합니다.
    return ",".join(updated_tag_list)

# 삭제할 태그들을 정의합니다. 여기서는 "Casual"과 "Great Soundtrack"을 삭제합니다.
tags_to_remove = "Casual,Great Soundtrack"

# "Tags" 열의 각 행에 대해 특정 태그들을 삭제합니다.
df["Tags"] = df["Tags"].apply(remove_tags, tags_to_remove=tags_to_remove)

# 변경된 데이터프레임을 CSV 파일로 내보냅니다.
df.to_csv("final_dataset_removed_tags.csv", index=False)
```

```python
# 중복된 항목을 제거하는 함수
def remove_duplicates(tags):
    # 태그들을 쉼표(,)로 분리하여 리스트로 변환합니다.
    tag_list = tags.split(",")
    # 중복을 제거합니다.
    unique_tags = list(set(tag_list))
    # 중복이 제거된 태그들을 다시 쉼표로 연결하여 하나의 문자열로 반환합니다.
    return ",".join(unique_tags)

# "Tags" 열의 각 행에 대해 중복된 항목을 제거합니다.
df["Tags"] = df["Tags"].apply(remove_duplicates)
```

```python
# 쉼표로 구분된 열 선택
columns_to_split = ['Genres', 'Tags']

# 쉼표로 나누어 리스트로 변환
for column in columns_to_split:
    df[column] = df[column].str.split(',')
```
